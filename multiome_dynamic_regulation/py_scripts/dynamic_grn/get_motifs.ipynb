## Jaspar

```python
from gimmemotifs.motif import read_motifs
import os, glob
from gimmemotifs.motif import MotifConfig
config = MotifConfig()
motif_dir = config.get_motif_dir()

# Get motif data names
motifs_data_name = [i for i in os.listdir(motif_dir) if i.endswith(".pfm")]
motifs_data_name.sort()

path = os.path.join(motif_dir, "JASPAR2020_vertebrates.pfm")
motifs_2 = read_motifs(path)
display(len(motifs_2))
```

```python
# save motifs_2 list of motifs in one single pfm file
with open('/ocean/projects/cis240075p/asachan/datasets/B_Cell/multiome_1st_donor_UPMC_aggr/dictys_outs/actb1_added_v2/data/JASPAR2020_vertebrates.pfm', 'w') as f:
    for i, motif in enumerate(motifs_2):
        # Add newline before motif name (except for first motif)
        if i > 0:
            f.write('\n')
        f.write(motif.to_pfm())
```

```python
# Check for CREB motifs
creb_motifs_2 = [motif for motif in motifs_2 if 'creb' in motif.id.lower()]

# Display the found motifs with their IDs and consensus sequences
for motif in creb_motifs_2:
    print(f"ID: {motif.id}")
    print(f"Consensus: {motif.consensus}")
```
